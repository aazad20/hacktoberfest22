```python
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
```

```python
df = pd.read_csv("train.csv")
```

```python
x_train = df.iloc[:,1:-1]
```

```python
x_train.columns
```

```python
y_train = df.iloc[:,-1:]
```

```python
y_train.columns
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.Sequential([
    keras.layers.Dense(units=50,input_shape=(3000,)),
    #keras.layers.Dense(units=50, activation='relu'),
    #keras.layers.Dense(units=40, activation='relu'),
    #keras.layers.Dense(units=40, activation='relu'),
    #keras.layers.Dense(units=30, activation='relu'),
    #keras.layers.Dense(units=30, activation='relu'),
    keras.layers.Dense(units=25, activation='relu',kernel_regularizer=keras.regularizers.l1()),
    keras.layers.Dense(units=15, activation='relu',kernel_regularizer=keras.regularizers.l2(0.001)),
    keras.layers.Dense(units=7, activation='relu',kernel_regularizer=keras.regularizers.l1_l2(0.01)),
    keras.layers.Dense(units=1, activation='sigmoid')
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=[tf.keras.metrics.BinaryAccuracy()])
```

```python
x_train = np.asarray(x_train)
y_train = np.asarray(y_train)
```

```python
model.fit( x=x_train,y=y_train,epochs=100,validation_split=0.30)
```

```python
prediction = model.predict(x_train)
prediction
```

```python
#prediction.astype(int)
model.evaluate(x_train,y=y_train)
```
